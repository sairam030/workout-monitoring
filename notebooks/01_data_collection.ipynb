**Objectives:**

2. Extract 33 body landmarks from videos

**Why MediaPipe?**
- Real-time performance (30+ FPS on CPU)
- 33 3D landmarks with visibility scores
- Robust to partial occlusion
- No GPU required
- Industry standard (used by Google Fit)

- `x`: Horizontal position [0, 1] normalized
- `y`: Vertical position [0, 1] normalized

## Step 3: Understanding Landmark Structure

**33 Landmarks (MediaPipe Pose):**
```
0: Nose                 17-22: Hands
1-10: Face contour      23-24: Hips
11-12: Shoulders        25-26: Knees
13-14: Elbows           27-28: Ankles
15-16: Wrists           29-32: Feet
```

````python
# Demonstrate landmark extraction (with dummy data)
dummy_landmarks = np.random.rand(132)  # Simulated landmarks

# Extract specific landmark (e.g., left elbow)
left_elbow_idx = 13
start = left_elbow_idx * 4
left_elbow_x = dummy_landmarks[start + 0]
left_elbow_y = dummy_landmarks[start + 1]
left_elbow_z = dummy_landmarks[start + 2]
left_elbow_vis = dummy_landmarks[start + 3]

print("Left Elbow Landmark:")
print(f"  x: {left_elbow_x:.3f} (horizontal position)")
print(f"  y: {left_elbow_y:.3f} (vertical position)")
print(f"  z: {left_elbow_z:.3f} (depth)")
print(f"  visibility: {left_elbow_vis:.3f} (confidence)")
````
